```python
df1=pd.read_excel(file_path+'Historical unit sales by SKU by time.xlsx')
df1
```

```python
df1['Average']=df1.mean(axis = 1)

df1.sort_values(by=['Average'])
```

```python
# importing required libraries
import seaborn as sns
sns.set()
sns.set(style="darkgrid")


import numpy as np
import pandas as pd

# importing matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
plt.rcParams['figure.figsize']=(10,10)
```

```python
sns.histplot(df1['Average'],bins=50)
```

```python
bins = [0,50,100,300]
labels = ['LowSales', 'MediumSales','HighSales']
df1['SalesDate'] = pd.cut(x=df1['Average'],bins=bins,labels=labels, include_lowest = True)
df1
```

```python
df2 = df1.groupby("SalesDate").mean()
df2
```

```python
df2 = df2.drop(labels=["Average"], axis=1)
df2
```

```python
df=df2.transpose()
df
```

```python
import datetime
```

```python
import warnings
import matplotlib.pyplot as plt
y = df['HighSales']
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(y,marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(y.resample('M').mean(),marker='o', markersize=8, linestyle='-', label='Monthly Mean Resample')
ax.set_ylabel('HighSales')
ax.legend();
```
